```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Unemployment Ohio.csv")
df = df.fillna(0)
df.head()
```

```python
df.describe()
```

```python
x_axis = np.arange(len(df))
tick_locations = [value for value in x_axis]
plt.bar(x_axis, df["Unemployment_Rate"], color = 'r', alpha = 0.5, align = 'center')
plt.xticks(tick_locations, df["County_Name"], rotation=45)
plt.title("Unemployment Rates in Ohio by County in 2019")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
```

```python
county19_df = df.iloc[[8, 17, 24]]
county19_df
```

```python
x_axis = np.arange(len(county19_df))
tick_locations = [value for value in x_axis]
plt.bar(x_axis, county19_df["Unemployment_Rate"], color = 'r', alpha = 0.5, align = 'center')
plt.xticks(tick_locations, county19_df["County_Name"], rotation=45)
plt.title("Unemployment Rates in Ohio by County in 2019")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
```

```python
ohio18_df = pd.read_csv("Unemployment_Ohio_2018.csv")
ohio18_df = ohio18_df.fillna(0)
ohio18_df.head()
```

```python
x_axis = np.arange(len(ohio18_df))
tick_locations = [value for value in x_axis]
plt.bar(x_axis, ohio18_df["Unemployment_Rate"], color = 'r', alpha = 0.5, align = 'center')
plt.xticks(tick_locations, ohio18_df["County_Name"], rotation=45)
plt.title("Unemployment Rates in Ohio by County in 2018")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
```

```python
combined_data = pd.merge(df, ohio18_df, on="County_Name")
combined_data.head()
```

```python
del combined_data['State_y']
combined_data = combined_data.rename(columns={"State_x":"State",
                                             "Labor_Force_x":"Labor_Force_19",
                                             "Employed_x":"Employed_19",
                                             "Unemployed_x":"Unemployed_19",
                                             "Unemployment_Rate_x":"Unemployment_Rate_19",
                                             "Labor_Force_y":"Labor_Force_18",
                                             "Employed_y":"Employed_18",
                                             "Unemployed_y":"Unemployed_18",
                                             "Unemployment_Rate_y":"Unemployment_Rate_18"})
combined_data.head()
```

```python
x_axis = combined_data["County_Name"]
county19 = plt.plot(x_axis, combined_data["Unemployment_Rate_19"], color="blue", label="2019")
county18 = plt.plot(x_axis, combined_data["Unemployment_Rate_18"], color="red", label="2018")
plt.title("Unemployment Rates in Ohio by County between 2018 and 2019")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
county1819_df = combined_data.iloc[[8, 17, 24]]
county1819_df
```

```python
x_axis = county1819_df["County_Name"]
county19 = plt.plot(x_axis, county1819_df["Unemployment_Rate_19"], color="blue", label="2019")
county18 = plt.plot(x_axis, county1819_df["Unemployment_Rate_18"], color="red", label="2018")
plt.legend()
plt.title("Unemployment Rates in Ohio by County between 2018 and 2019")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
