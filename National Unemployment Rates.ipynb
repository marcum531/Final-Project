```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("new_employment.csv")
df = df.fillna(0)
df = df.replace(to_replace = "*",value = "0")
df = df.replace(to_replace = "(X)", value = "0")
df.head()
```

```python
# x_axis = df["State"]
state15 = plt.plot(df["2015_estimate_employment_civilian_labor_force_unemployment_rate"], color="black", label="2015")
state16 = plt.plot(df["2016_estimate_employment_civilian_labor_force_unemployment_rate"], color="green", label="2016")
state17 = plt.plot(df["2017_estimate_employment_civilian_labor_force_unemployment_rate"], color="pink", label="2017")
state18 = plt.plot(df["2018_estimate_employment_civilian_labor_force_unemployment_rate"], color="blue", label="2018")
state19 = plt.plot(df["2019_estimate_employment_civilian_labor_force_unemployment_rate"], color="red", label="2019")
plt.legend()
plt.title("Unemployment Rates in the US between 2015 through 2019")
plt.xlabel("States")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
