```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Unemployment Ohio.csv")
df = df.fillna(0)
df.head()
```

```python
df.describe()
```

```python
x_axis = np.arange(len(df))
tick_locations = [value for value in x_axis]
```

```python
plt.bar(x_axis, df["Unemployment_Rate"], color = 'r', alpha = 0.5, align = 'center')
plt.xticks(tick_locations, df["Unemployment_Rate"], rotation=45)
plt.title("Unemployment Rates in Ohio by County in 2019")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
```

```python
county_df = df.iloc[[8, 17, 24]]
county_df
```

```python
x_axis = np.arange(len(county_df))
tick_locations = [value for value in x_axis]

plt.bar(x_axis, county_df["Unemployment_Rate"], color = 'r', alpha = 0.5, align = 'center')
plt.xticks(tick_locations, county_df["County_Name"], rotation=45)
plt.title("Unemployment Rates in Ohio by County in 2019")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
```

```python
combined_data = pd.merge(df, Ohio18_df, on="County_Name")
combined_data.head()
```

```python
del combined_data['State_y']
combined_data = combined_data.rename(columns={"State_x":"State", 
                                             "Labor_Force_x": "Labor_Force_19",
                                             "Employed_x":"Employed_19",
                                             "Unemployed_x":"Unemployed_19",
                                              "Unemployment_Rate_x":"Unemployment_Rate_19",
                                             "Labor_Force_y":"Labor_Force_18",
                                             "Employed_y":"Employed_18",
                                             "Unemployed_y":"Unemployed_18",
                                             "Unemployment_Rate_y":"Unemployment_Rate_18"})
combined_data.head()
```

```python
x_axis = combined_data["County_Name"]
ohio18 = plt.plot(x_axis, combined_data['Unemployment_Rate_18'], color="blue", label="2018")
ohio19 = plt.plot(x_axis, combined_data['Unemployment_Rate_19'], color="red", label="2019")
plt.tight_layout()
```
