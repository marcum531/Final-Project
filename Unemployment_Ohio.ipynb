```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Unemployment Ohio.csv")
df = df.fillna(0)
df.head()
```

```python
df.describe()
```

```python
new_df = df
for row in new_df:
    
    if  != 'Butler County' | 'Cuyahoga County' | 'Franklin County':
        new_df.drop[row]

new_df.head()
```

```python
new_df[new_df.County_Name == 'Butler County' ]

new_df.head()
```

```python
new_df = df.loc[df['County_Name'].isin(['Butler County', 'Cuyahoga County', 'Franklin County'])]
```

```python
new_df.head()
```

```python
x_axis = np.arange(len(new_df))
tick_locations = [value for value in x_axis]
```

```python
plt.bar(x_axis, new_df["Unemployment_Rate"], color = 'r', alpha = 0.5, align = 'center')
plt.xticks(tick_locations, new_df["County_Name"], rotation=45)
plt.title("Unemployment Rates in Ohio by County")
plt.xlabel("County Name")
plt.ylabel("Unemployment Rate (%)")
plt.tight_layout()
```
